```python
#Importing Country Dataset
df_country = pd.read_excel("H:\\Hrushikesh\\Projects\\Exploratory Data Analysis\\Zomato Sales Analysis\\Country-Code.xlsx")
```

```python
# Merging the df and df_country where country codes are same
final_df = pd.merge(df,df_country,on = 'Country Code',how = 'left')
```

## Visualizing top 3 Countries that uses zomato

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# Top 3 countries that uses zomato 
plt.figure(figsize=(5,8))
plt.title("Top 3 Countries That Uses Zomato")
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%',labeldistance=1.1)
plt.show()
```

### Insight
##### Zomato having maximum records in India after United States and United Kingdom with 94.39%

```python
# finding the count of rating 
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# Renaming the count column
rating = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

## Insights
- When rating is between 4.5 to 4.9 it means Excellent
- When rating is between 4.0 to 4.4 it means Very Good
- When rating is between 3.5 to 3.9 it means Good
- When rating is between 2.5 to 3.4 it means Avearge
- When rating is between 2.0 to 2.4 it means Poor 

```python
rating.head()
```

```python
# Visualizing Aggregate rating
plt.figure(figsize=(12,6))
plt.title("Aggregate Rating Count")
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating,palette='muted')
plt.show()
```

##Insights
- 0.0 is the highest rating
- Maximum ratings are lies between 2.7 to 3.9
- 3.1 and 3.2 are the highest ratings
